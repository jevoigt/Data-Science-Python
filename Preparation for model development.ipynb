## We can now manage these datasets further by merging the two and forming training and test datasets.

```python
carpet['SetNumber'] = pd.Series([0 for x in range(len(carpet.index))])
```

```python
carpet.head()
```

```python
hardwood["SetNumber"] = pd.Series([1 for x in range(len(carpet.index))])
```

```python
hardwood.head()
```

```python
carwood = pd.concat([carpet, hardwood], axis=0)
```

```python
carwood.drop(["Number"], axis = 1, inplace = True)
```

```python
carwood.head()
```

```python
carwood.tail()
```

## Above we have created a new column in each dataset, called "SetNumber". This helps us identify which dataset any given observation belongs to. This becomes useful in the next step, where we merged our datasets into one, titled Carwood.

```python
randcarwood = carwood.sample(frac=1).reset_index(drop=True)
```

```python
randcarwood.head(10)
```

```python
randcarwood.tail(10)
```

```python
randcarwood.to_csv('randcarwood.csv')
```

## We have now merged our two datasets, shuffled the observations randomly, and exported this new dataframe as a csv file titled "randcarwood.csv". This allows us to perform any analysis on just one dataset and being able to pull either carpet or hardwood data using our SetNumber column.

## We will now seperate this new dataframe into two new sets, one will be our "training" data set and the other will be our "test" data set. We can eventually us our "training" data set in developing statistical models to help us make predicitions or perform other analysis. Our "test" dataset acts as a new sample that we can then apply to our model and test to see if it performs well for both sets of data.

```python
train, test = train_test_split(randcarwood, test_size=0.2, shuffle=True)
```

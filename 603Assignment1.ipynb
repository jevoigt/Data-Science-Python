Cleaning: 

Before we begin cleaning our data, we must analyze our data to know where we need to make corrections. We can take a look at each variable individually. The variable "symboling" has many values that include the character "?", which is not useful to us. The variable "engine-location" has many values that also include characters we do not need, such as "[", "]", "?", ",", and repeated words like "front". We want to correct these characters in order to make our variable more uniform. We know that "front," and "front", probably mean the same thing, so we need to correct these typos so we can organize the data more effectively.

```python
#Removing unwanted characters from "engine-location"
unwanted_characters = ['[', ',', '-', ']', '.', '?']

def clean_dates(item):
    dop= str(item.loc['engine-location'])
    
    if dop == 'nan' or dop[0] == '[':
        return np.NaN
    if dop == 'font' or dop[0] == ',' or dop[0]=='?':
        return 'front'
    
    for character in unwanted_characters:
        if character in dop:
            character_index = dop.find(character)
            dop = dop[:character_index]
    
    return dop

df['engine-location'] = df.apply(clean_dates, axis = 1)
df.head(10)
```

Above is a"dirty" column that we decided to clean. We removed all of the unwanted characters from all the values of "engine-location" and made the values of each row more uniform. We changed 3 of the variables within this column to better suit the actual engine locations we are looking for. You can see below the end result of the changes.

```python
df['engine-location'].value_counts() #Frequency count of our values in "engine-location"
```

Above we can see the modified value counts for the column "engine-location". We have changed the names of anything that resembled "front" to front so now we have 202 observations under "front" and 3 observations under "rear".

Transforming:

We can tranform our dataset values, rows, and columns for many different reasons and in lots of different ways. We can tranform our data to make it more uniform and organized, which can make it easier on us, the user, to inperpret, and also easier for the computer to process in analyses. This can also make comparing two or more groups, as we can visualize them on a different scale. A log transformation of the data is common for this. Transforming your data can also make it more universal, and it can then be applied to different applications and other things.

One way we could tranform this data set is by tranforming some of the column names. If we take a look at the columns "wheel-base", "length", "width", and "height", we can see they all have numeric values, but no units of measurement. Due to our prior knowledge, and doing a little bit of research with the information in our data set, it is safe to assume that the units of measurment here are in inches (many websites list details about these makes in this unit). We can tranform these column names to give analyists a better understanding of how big these measurements really are. 

```python
new_names =  {'wheel-base': 'wheel-base (inches)', #Renaming (transforming) column names
              'length': 'length (inches)',
              'width': 'width (inches)',
              'height': 'height (inches)'}

df.rename(columns = new_names, inplace = True)
df.head()
```

You can see from the head of the data above, We have now changed the column names to give readers or analysis a better understanding of what they are reading. This adds a scale to these columns and makes them easier to interpret. 

```python
engine_size = df["engine-size"]            
wheel_base = df["wheel-base (inches)"]
plt.hist(wheel_base, bins = 10, color = 'teal', \
           alpha = 0.5)
plt.hist(engine_size, bins = 20, color = 'blue', \
            alpha = 0.5)
plt.show()
```

Here we have our histogram of our two columns "engine-size" and "wheel-base (inches)". Both of these columns appear to have very similar means, and both seem to be right-skewed. The deviation of the engine size column is far greater than that of the wheel base column. The wheel base column has a realtively low deviation from the mean.

```python
df['natural_log_ES'] = np.log(df['engine-size'])
df['natural_log_WB'] = np.log(df['wheel-base (inches)'])
Log_ES = df['natural_log_ES']
Log_WB = df['natural_log_WB']
plt.hist(Log_WB, bins = 10, color = 'teal', \
           alpha = 0.5)
plt.hist(Log_ES, bins = 20, color = 'blue', \
            alpha = 0.5)
plt.show()
```

This new histogram shows both of our columns, but after we have applied the log transformation. This tranformation is typically applied to two very different plots, or plots that are very skewed one way or another. After the log tranformation, we can see both of these graphs on a differnt scale that makes them a little more comparable visually. While this plot could probably be modified further (changing the base of our log or adding labels to further improve visualization) I would choose the logged plot over the previous plot, because these two distributions are far more normal, and also have less deviation. This makes them much easier to overlap and analyze.
